```julia
include("include.jl");
```

```julia
data = MyOriginalPortfolioDataSet()
```

```julia
original_dataset = data["dataset"]
```

```julia
maximum_number_trading_days1 = original_dataset["AAPL"] |> nrow
```

```julia
dataset1 = Dict{String,DataFrame}();
for (ticker,data) ∈ original_dataset
    if (nrow(data) == maximum_number_trading_days1)
        dataset1[ticker] = data;
    end
end
```

```julia
dataset1
```

```julia
oos_data = MyOutOfSamplePortfolioDataSet()
```

```julia
oos_dataset = oos_data["dataset"]
```

```julia
maximum_number_trading_days2 = oos_dataset["AAPL"] |> nrow
```

```julia
dataset2 = Dict{String,DataFrame}();
for (ticker,data) ∈ oos_dataset
    if (nrow(data) == maximum_number_trading_days2)
        dataset2[ticker] = data;
    end
end
dataset2
```

```julia
# Find tickers that are in dataset1 but not in dataset2
diff_tickers = setdiff(keys(dataset2), keys(dataset1))

# Print the list
println(diff_tickers)
```
